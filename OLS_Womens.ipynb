```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
data = pd.read_excel("WomensHundred_Dummy.xlsx")
data.head()
```

```python
data.info()
```

### Splitting data : Pace & Spin

```python
df = data[["Date", "BatterID", "BowlerID", "PitchX", "PitchY", "ReleaseY", "ReleaseZ", "PastY", "PastZ", "Bowler Type", "Runs"]].copy()
df_pace = df[df["Bowler Type"].str.endswith(('M', 'F'))].copy()
df_spin = df[df["Bowler Type"].str.endswith(('B'))].copy()
```

```python
features = ["PitchX","PitchY","ReleaseY", "ReleaseZ", "PastY", "PastZ"]
```

### OLS Regression for Pace Data

```python
Y_pace = df_pace["Runs"]
X_pace = sm.add_constant(df_pace[features])
res_pace = sm.OLS(Y_pace, X_pace).fit()
res_pace.summary()
```

### OLS Regression For Spin Data

```python
Y_spin = df_spin["Runs"]
X_spin = sm.add_constant(df_spin[features])
res_spin = sm.OLS(Y_spin, X_spin).fit()
res_spin.summary()
```

```python
df_pace["expected_runs"] = res_pace.predict(X_pace)
df_pace["residual"] = df_pace["Runs"] - df_pace["expected_runs"]
```

### Aggregation For Pace Batters

```python
pace_batters = (
    df_pace
    .groupby("BatterID")
    .agg(
        n_balls=("Runs", "size"),
        mean_residual=("residual", "mean"),
        std_residual=("residual", "std")
    )
)
# se is "Standart Error"
pace_batters["se"] = pace_batters["std_residual"] / pace_batters["n_balls"]**0.5
pace_batters["t_stat"] = pace_batters["mean_residual"] / pace_batters["se"]
```

# -----------------------------------------------------

```python
df_spin["expected_runs"] = res_spin.predict(X_spin)
df_spin["residual"] = df_spin["Runs"] - df_spin["expected_runs"]
```

### Aggregation For Spin Batters

```python
spin_batters = (
    df_spin
    .groupby("BatterID")
    .agg(
        n_balls=("Runs", "size"),
        mean_residual=("residual", "mean"),
        std_residual=("residual", "std")
    )
    .dropna()
)
```

```python
spin_batters["se"] = spin_batters["std_residual"] / (spin_batters["n_balls"] ** 0.5)
spin_batters["t_stat"] = spin_batters["mean_residual"] / spin_batters["se"]
```
